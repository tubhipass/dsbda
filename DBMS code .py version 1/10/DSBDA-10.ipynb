```python
plt.figure(figsize = (10, 7))
x = info["SepalLengthCm"]

plt.hist(x, bins = 20, color = "green")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (10, 7))
x = info["SepalWidthCm"]

plt.hist(x, bins = 20, color = "green")
plt.title("Sepal Width in cm")
plt.xlabel("Sepal_Width_cm")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (10, 7))
x = info["PetalLengthCm"]

plt.hist(x, bins = 20, color = "green")
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize = (10, 7))
x = info["PetalWidthCm"]

plt.hist(x, bins = 20, color = "green")
plt.title("Petal Width in cm")
plt.xlabel("Petal_Width_cm")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize = (8, 4))
info.boxplot()
plt.show()
```

```python
fig, axes=plt.subplots(2, 2, figsize=(16, 8))
axes[0,0].set_title("Distribution of sepal length")
axes[0,0].hist(info["SepalLengthCm"]);

axes[0,1].set_title("Distribution of sepal width")
axes[0,1].hist(info["SepalWidthCm"]);

axes[1,0].set_title("Distribution of petal length")
axes[1,0].hist(info["PetalLengthCm"]);

axes[1,1].set_title("Distribution of petal width")
axes[1,1].hist(info["PetalWidthCm"]);
plt.show()
```

```python
fig, axes=plt.subplots(2, 2, figsize=(16, 9))
axes[0,0].set_title("Distribution of sepal length")
sns.boxplot( y="SepalLengthCm", x="Species", data=info, orient='v', ax=axes[0,0])
axes[0,1].set_title("Distribution of sepal width")
sns.boxplot( y="SepalWidthCm", x="Species", data=info, orient='v', ax=axes[0,1])
axes[1,0].set_title("Distribution of petal length")
sns.boxplot( y="PetalLengthCm", x="Species", data=info, orient='v', ax=axes[1,0])
axes[1,1].set_title("Distribution of petal width")
sns.boxplot( y="PetalWidthCm", x="Species", data=info, orient='v', ax=axes[1,1])
plt.show()
```
